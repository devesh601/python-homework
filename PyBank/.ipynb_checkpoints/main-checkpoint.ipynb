```python
# created Github Repo python-homework
# created PyBank and PyRamen directories 
# created main.ipynb in PyBank 
# Pushed the changes using Git Bash
# $ git push origin
# Enumerating objects: 8, done.
# Counting objects: 100% (8/8), done.
# Delta compression using up to 8 threads
# Compressing objects: 100% (5/5), done.
# Writing objects: 100% (6/6), 867 bytes | 867.00 KiB/s, done.
# Total 6 (delta 0), reused 0 (delta 0), pack-reused 0
# To https://github.com/devesh601/python-homework.git
#   27b0ea4..e7e066e  main -> main

# Importing file

import pandas as pd 
from pathlib import Path 

budget_data = pd.read_csv(r'C:\Users\deves\python-homework\PyBank\budget_data.csv')

# print(budget_data)

# creating empty lists to move budget_date.csv columns

budget_date = []
budget_profit_loss = []

# converting column data to list 

budget_date = budget_data['Date'].tolist()
budget_profit_loss = budget_data['Profit/Losses'].tolist()

# print(budget_date)
# print(budget_profit_loss)

# creating result format 

print("Financial Analysis")
print("-------------------------------")

# the total number of months included in the dataset - can be calulated by using the lenght of budget_date list

# adding the value lenth to a variable for printing later

t_months = len(budget_date)
print(f'Total Months : {t_months}')

# the net total amount of profit/losses over the entire period can calculate by using a for loop and adding all the values of list- budget_profit_loss 

# defining variable to add values 

total_profit_loss = 0

for i in range(0,len(budget_profit_loss)):
    total_profit_loss = total_profit_loss + budget_profit_loss[i]
print(f'Total: ${total_profit_loss}')

# change of profit and loss can taken in to a new list using a for loop and average can be calculated 
diff = 0 
change_profit_loss = []

for i in range(0,(len(budget_profit_loss)-1)):
    diff = budget_profit_loss[(i+1)] - budget_profit_loss[i]
    change_profit_loss.append(diff)
# print(change_profit_loss)
# calculating average 
total_change = 0
for i in range(0,len(change_profit_loss)):
    total_change = total_change + change_profit_loss[i]
# print(total_change)
average = total_change / len(change_profit_loss)
average = round(average,2)
print(f'Average Change: ${average}')

# find the max value in the change_profit_loss list and its index using for loop 

max_change_value = 0 
for i, value in enumerate(change_profit_loss):
    if value>max_change_value:
        max_change_value = value
        index = i
# print(max_change_value)
# since change_profit_list starts at index 1 of budget_date, finding out the month based on highest change value's index
max_month = budget_date[(index+1)]
print(f'Greatest Increase in Profits: {max_month} (${max_change_value})')

# find the lowest value in the change_profit_loss list and its index using for loop 

min_change_value = 0
for i, value_min in enumerate(change_profit_loss):
    if value_min < min_change_value:
        min_change_value = value_min
        index_min = i
# print(min_change_value)
# since change_profit_list starts at index 1 of budget_date, finding out the month based on highest change value's index
min_month = budget_date[(index_min+1)]
print(f'Greatest Decrease in Profits: {min_month} (${min_change_value})')


#Exporting results to a text file

import sys 

stdoutOrigin=sys.stdout 
sys.stdout = open("log.txt", "w")

print("Financial Analysis")
print("-------------------------------")
print(f'Total Months : {t_months}')
print(f'Total: ${total_profit_loss}')
print(f'Average Change: ${average}')
print(f'Greatest Increase in Profits: {max_month} (${max_change_value})')
print(f'Greatest Decrease in Profits: {min_month} (${min_change_value})')

sys.stdout.close()
sys.stdout=stdoutOrigin
```
